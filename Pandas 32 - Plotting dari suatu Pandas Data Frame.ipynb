## Persiapan Data Frame
- n_rows untuk menentukan jumlah baris yang ingin kita buat
- n_cols untuk menentukan jumlah kolom yang yang ingin kita buat
- randint untuk menampilkan nilai secara acak berdasarkan jumlah yang ditentukan

```python
n_rows = 40
n_cols = 5
cols = tuple ('ABCDE')

df = pd.DataFrame(np.random.randint(1, 20, size=(n_rows, n_cols)),
                  columns=cols)
df
```

## Line Plot

- membuat parameter kind dengan nilai line

```python
df.head()
```

```python
df.plot(kind='line')
```

```python
df[['A', 'B']].plot(kind='line')
```

## Barplot

- memberi nilai pada parameter kind (bar)

```python
df.head()
```

```python
df.plot(kind='bar')
```

```python
df[['A', 'B']].plot(kind='bar')
```

```python
df[['A', 'B']].head().plot(kind='bar')
```

```python
df[['A', 'B']].head().plot(kind='barh')
```
