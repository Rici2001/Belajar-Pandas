#02: Pemilihan baris (rows selection) pada Data Frame

Hal-Hal yang diperlukan, antara lain:
1) Membuat Data Frame

```python
import pandas as pd
import numpy as np

print(pd.__version__)
print(np.__version__)
```

```python
n_rows = 15
n_cols = 4
cols = tuple('RICI')

df = pd.DataFrame(np.random.randint(1, 10, size=(n_rows, n_cols)),
                  columns=cols)
df
```

#Bisa kita liat ini hasil dari inputan kita, nilai nilai nya selalu berbeda karena ini nilainya random.

---------------------------------------------------
SELECTION DENGAN OPERATOR LOGIKA |

#Tanda | bisa kita artikan sebagai or ( atau ), Biar lebih jelas silahkan melihat contoh dibawah ini :

```python
df[(df['R'] == 1) | (df['R'] == 5)]
```

#Maksud dari hasil output ini sangat mudah bisa kita pahami, secara logika kita memasukan nilai operator logikanya angka 1 | 5 di kolom R.

#Bisa juga kita menggunakan cara yang lebih sederhana dengan memanfaatkan fungsi isin(). Perhatikan di bawah ini .

```python
df[df['R'].isin([1, 5])]
```

#Berikut ini adalah hasil output dengan mengunakan fungsi isin. Dan hasilnya sama persis dengan fungsi sebelumnya.

------------------------------------------------------------
MENGGUNAKAN OPERASI NEGASI (~)
#Digunakan untuk menampilkan nilai yang dari kolom yang kita pilih tidak sama dengan dengan nilai yang kita input 1 | 5 (Lihat di contoh soal sebelumnya)

```python
df[~df['R'].isin([1, 5])]
```

#Berikut adalah output dari kolom R yang tidak sama dengan angka yang kita input (1 | 3) Dengan arti lain, selain angka tersebut akan di tampilkan di output.
